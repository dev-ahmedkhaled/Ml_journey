# liner_regression from scratch numpy only
Data set used is Salary Prediction data from kaggle: https://www.kaggle.com/datasets/krishnaraj30/salary-prediction-data-simple-linear-regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('Salary_dataset.csv')

print(data.head())

X=data['YearsExperience'].values
Y=data['Salary'].values

# lets do a scatter plot to see data distribution
plt.scatter(X,Y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
plt.show()

```

as we can see linear data distribution 

```python

# Mean Squared Error func to evaluate 
def Cost_function(X,Y,w,b):
    m=len(X)
    cost_sum= 0
    for i in range(m):
        f= w*X[i]+b
        cost=(f-Y[i])**2
        cost_sum+=cost
    total_cost = (1/(2*m))*cost_sum
    return total_cost

def Gradient_descent(x,y,learn_rate,number_interations):
    w=0
    b=0
    for i in range(number_interations):
        dc_dw,dc_db=Gradeint_function(x,y,w,b)

        w=w-learn_rate*dc_dw
        b=b-learn_rate*dc_db

        print(f"interation: {i} ,Cost :{Cost_function(x,y,w,b)}")

    return w,b
    

def Gradeint_function(x,y,w,b):
    m=len(x)
    dc_dw=0
    dc_db=0
    for i in range(m):
        f= w*x[i]+b
        dc_dw=+(f-y[i])*x[i]
        dc_db+=(f-y[i])

    dc_dw=(1/m)*dc_dw
    dc_db=(1/m)*dc_db

    return dc_dw,dc_db
```

```python
learning_rate= 0.01
iterations=10000

final_w,final_b=Gradient_descent(X,Y,learning_rate,iterations)

print(f"final w: {final_w}, final b : {final_b}")
```

```python
plt.scatter(X,Y, label="data points")

x_vals=np.linspace(min(X),max(X),100)
y_vals=final_w*x_vals+final_b
plt.plot(x_vals,y_vals,color='red',label="regression line")
plt.xlabel('user experince')
plt.ylabel('salary')
plt.legend()
plt.show()
```
